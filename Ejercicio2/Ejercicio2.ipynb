```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import scipy as scipy
from math import *
from tqdm import *
```

```python
t_f = 1000
n = 100000
```

```python
def r():
    aux = int(np.round(np.random.uniform(0,1)))
    if aux == 0:
        r = 0.5
    else:
        r = 2
    return(r)
```

```python
def number(t):
    f = np.random.rand()
    for j in range(t-1):
        f = f*r()
    return(f)
```

```python
his = np.zeros(n)
aux = np.zeros(n)

for i in tqdm(range(n)):
    aux[i]=number(t_f)
    his[i]=np.fix(abs(aux[i])*10**(-np.floor(log10(abs(aux[i])))))
```

```python
bins=[0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5]
points=np.linspace(1,9,9)
bl = np.array([0.301,0.176,0.125,0.097,0.079,0.067,0.058,0.051,0.046])*n
plt.hist(his,bins)
plt.plot(points,bl,'.',label='Ley de Benford')
plt.xlim([0.5,9.5])
plt.xlabel('1ª cifra significativa')
plt.legend()
plt.savefig('ejercicio2')
plt.show()
```
