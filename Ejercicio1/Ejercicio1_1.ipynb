```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from math import *
from tqdm import *
```

```python
data = open('ejercicio1_1.txt','r')
aux = data.readlines()
n = len(aux)
mun_gran = np.zeros(n)

for i in range(n):
    aux[i] = aux[i].split()
    mun_gran[i] = aux[i][0]

data.close()
```

```python
his = np.zeros(n)

for i in tqdm(range(n)):
    his[i]=np.fix(abs(mun_gran[i])*10**(-np.floor(log10(abs(mun_gran[i])))))
```

```python
bins=[0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5]
points=np.linspace(1,9,9)
bl = np.array([0.301,0.176,0.125,0.097,0.079,0.067,0.058,0.051,0.046])*np.size(mun_gran)
plt.hist(his,bins)
plt.plot(points,bl,'.',label='Ley de Benford')
plt.xlim([0.5,9.5])
plt.xlabel('1ª cifra significativa')
plt.legend()
plt.savefig('ejercicio1_1')
plt.show()
```

```python
c=1
plt.hist(np.log10(mun_gran*c),20,range=[np.log10(min(mun_gran*c)),np.log10(max(mun_gran*c))])
plt.show()
```
